```python
from collections import deque

# Verifica se o estado é seguro, ou seja, não há mais canibais do que missionários em nenhuma das margens
def estado_valido(missionarios, canibais):
    if missionarios >= 0 and missionarios <= 3 and canibais >= 0 and canibais <= 3:
        if missionarios == 0 or missionarios >= canibais:
            return True
    return False

# Função que realiza a busca em largura para encontrar a solução
def bfs_missionarios_canibais():
    estado_inicial = (3, 3, 1, 0, 0, 0)  # (missionários_iniciais, canibais_iniciais, barco na margem inicial, missionários_finais, canibais_finais, barco na margem final)
    fila = deque([estado_inicial])
    visitados = set([estado_inicial])
    pais = {estado_inicial: None}

    primeiro_movimento = True  # Variável para controlar o primeiro movimento

    while fila:
        estado_atual = fila.popleft()

        # Se todos os missionários e canibais estiverem na margem final, encontramos a solução
        if estado_atual[3] == 3 and estado_atual[4] == 3:
            caminho = []
            while estado_atual:
                caminho.append(estado_atual)
                estado_atual = pais[estado_atual]
            return caminho[::-1]  # Retorna o caminho da solução

        # Gerar os próximos estados possíveis
        miss_inicial, can_inicial, barco_inicial, miss_final, can_final, barco_final = estado_atual

        # Definir os movimentos baseados na posição do barco
        if barco_inicial == 1:  # Barco na margem inicial
            if primeiro_movimento:
                movimentos = [(1, 1)]  # Primeiro movimento deve ser de 1 missionário e 1 canibal
                primeiro_movimento = False  # Desabilitar o controle do primeiro movimento após a primeira travessia
            else:
                movimentos = [(1, 0), (2, 0), (0, 1), (0, 2), (1, 1)]  # Possíveis combinações após o primeiro movimento
        else:  # Barco na margem final
            movimentos = [(-1, 0), (-2, 0), (0, -1), (0, -2), (-1, -1)]  # Movimentos reversos ao barco voltar

        for m, c in movimentos:
            novo_miss_inicial = miss_inicial - m
            novo_can_inicial = can_inicial - c
            novo_miss_final = miss_final + m
            novo_can_final = can_final + c
            novo_barco_inicial = 1 - barco_inicial

            novo_estado = (novo_miss_inicial, novo_can_inicial, novo_barco_inicial, novo_miss_final, novo_can_final, 1 - novo_barco_inicial)

            if estado_valido(novo_miss_inicial, novo_can_inicial) and estado_valido(novo_miss_final, novo_can_final):
                if novo_estado not in visitados:
                    visitados.add(novo_estado)
                    fila.append(novo_estado)
                    pais[novo_estado] = estado_atual

    return None  # Se nenhuma solução for encontrada

# Imprimir a solução
solucao = bfs_missionarios_canibais()
if solucao:
    print("Solução encontrada!")
    for passo in solucao:
        print(passo)
else:
    print("Nenhuma solução encontrada.")
```
